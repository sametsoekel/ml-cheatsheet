```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset("diamonds")
kategori=["Fair","Good","Very Good","Premium","Ideal"]
df.cut=df.cut.astype(pd.api.types.CategoricalDtype(categories=kategori,ordered=True))
df.cut
```

```python
(df.cut
.value_counts().plot.barh().set_title("Çatışmaya Devam"));
```

```python
sns.barplot(x="cut",y=df.cut.index, data=df);
```

```python
(sns
 .FacetGrid(df,
               hue="cut",
               height=5,
               xlim=(0,10000))
 .map(sns.kdeplot,"price",shade=True)
 .add_legend());
```
