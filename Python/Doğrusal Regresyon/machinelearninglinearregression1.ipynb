```python
#Basit doğrusal regresyon uygulaması

import pandas as pd
import seaborn as sns
```

```python
ad=pd.read_csv("Advertising.csv",usecols=[1,2,3,4])
df = ad.copy()
```

```python
sns.pairplot(df, kind="reg");
```

```python
sns.jointplot(x="TV",y="sales",data=df,kind="reg");
```

```python
#MODEL GELİYOR BİSMİLLAH#

#Statsmodels ile modelleme
```

```python
import statsmodels.api as sm

X=df["TV"]
```

```python
X = sm.add_constant(X) #MATRİS İŞLEMLERİ İÇİN 1 SÜTUNU EKLEDİK
#Bağımsız değişen hazır ! :)
```

```python
y=df["sales"] #bağımlı değişken
```

```python
#lm lineer model demek

lm=sm.OLS(y,X)
```

```python
model = lm.fit()
```

```python
model.summary()
```
